```python
import networkx as nx
import matplotlib.pyplot as plt

grafo = nx.DiGraph()
grafo.add_edge('A', 'B')
grafo.add_edge('A', 'C')
grafo.add_edge('A', 'D')
grafo.add_edge('D', 'B')
grafo.add_edge('J', 'B')
grafo.add_edge('J', 'C')

print(nx.info(grafo))
```

```python
plt.figure(figsize=(6,6))
plt.axis('off')

import numpy as np
np.random.seed(0)

nx.draw_networkx(grafo)
```

```python
data = nx.read_edgelist('./facebook_combined.txt')
print(nx.info(data))
```

```python
plt.figure(figsize=(12,12))

position = nx.spring_layout(data)

nx.draw_networkx(data, node_size=20, pos=position, node_color='red', with_labels=False)
```

```python
from community import community_louvain

partitions = community_louvain.best_partition(data)
values = list(partitions.values())
print(partitions)
```

```python
plt.figure(figsize=(10,10))

nx.draw_networkx(data, node_size=20, pos=position, node_color=values, with_labels=False)
```
